```python
import d2lzh as d2l
```

```python
from mxnet import autograd, nd

def xyploy(x_vals, y_vals, name):
    d2l.set_figsize(figsize=(5,2.5))
    d2l.plt.plot(x_vals.asnumpy(), y_vals.asnumpy())
    d2l.plt.xlabel('x')
    d2l.plt.ylabel(name + '(x)')
    
    
```

```python
x = nd.arange(-8.0, 8.0, 0.1)
x.attach_grad()
with autograd.record():
    y = x.relu()
    y2 = x.tanh()
    y3 = x.sigmoid()
    
xyploy(x,y,'relu')    
xyploy(x,y2,'relu')
xyploy(x,y3,'relu')
```
